```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.array([1, 2, 3, 4, 5, 6])
ys = np.array([5, 4, 6, 5, 6, 7])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
xs = xs.reshape((6, 1))
clf.fit(xs, ys)
print(xs)
```

```python
clf.predict([[0], [7]])
```

```python
plt.plot([0, 7], [4, 7])

plt.scatter(xs, ys)
plt.show()
```

```python
from sklearn import datasets
import pandas as pd
```

```python
boston = datasets.load_boston()
df = pd.DataFrame(boston.data)        #create a dataframe from the data
df.head()
```

```python
print(boston.feature_names)     # gives the names of the features
print(boston.target[0:5])
```

```python
df.columns = boston.feature_names
print(df.head())
```

```python
from sklearn import cross_validation as cv        #split data into train and test
X_train, X_test, Y_train, Y_test = cv.train_test_split(df, boston.target, test_size = 0.2)     # (data, labels, percentage)
```

```python
clf2 = LinearRegression()
clf2.fit(X_train, Y_train)
```

```python
predictions = clf2.predict(X_test)
print(predictions)
clf2.score(X_test, Y_test)
```
